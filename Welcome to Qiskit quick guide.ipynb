### Imports 

```python
# from package import *  How to import everything from a python package 

from qiskit.visualization import * # bloch sphere graphs, bar graphs for measurements 
from qiskit import * # important functionality like QuantumCircuit() and etc

S_simulator=Aer.backends(name='statevector_simulator')[0] # allows you to simulate states
M_simulator=Aer.backends(name='qasm_simulator')[0] # allows you to simulate measurements 

# Shift + tab on functionality from a package to pull up documentation. 
```

```python
# make QuantumCircuit Object
my_quantum_circuit=QuantumCircuit(2,2)

#Apply H gate to first qubit
my_quantum_circuit.h(0) # counting of our qubits starts with 0 (apply hadamard gate onto first qubit)
my_quantum_circuit.cx(0,1)
#draw circuit
my_quantum_circuit.draw(output='mpl') # for .draw(output='mpl') must download matplotlib first, otherwise just use .draw()
```

```python
# measurements 

# measure the circuit! 
# my_quantum_circuit.measure(0,0) # measure the first qubit, store result in the first cbit 
# my_quantum_circuit.measure(1,1) 

my_quantum_circuit.measure([0,1], [0,1])

job = execute(my_quantum_circuit, M_simulator, shots=1024)
result = job.result()

counts = result.get_counts()  # this gives me my results

# visualizing results 

plot_histogram(counts)

# As you can see we create a bell state! 
```
